```python
num_of_tweets = 200
tweets = []
likes = []
time = []

for i in tweepy.Cursor(api.search, q="$GME", tweet_mode ="extended").items(num_of_tweets):
    tweets.append(i.full_text)
    likes.append(i.favorite_count)
    time.append(i.created_at)
    
wsb_df = pd.DataFrame({
    'tweets': tweets
    ,'likes': likes
    ,'time': time
})
wsb_df
```
